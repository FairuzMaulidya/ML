```python
import pandas as pd

data = pd.DataFrame({
    'X': [1, 2, 3, 8, 9, 10],
    'Y': [1, 2, 3, 8, 9, 10]
})

print(data)
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['X'], data['Y'])
plt.title('Data Awal')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(data[['X', 'Y']])

data['Cluster'] = kmeans.labels_

print(data)
```

```python
plt.scatter(data['X'], data['Y'], c=data['Cluster'], cmap='viridis')
plt.title('Hasil K-Means Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
print(f"Inertia (SSE): {kmeans.inertia_}")
```

```python
from sklearn.datasets import make_blobs

data, labels = make_blobs(n_samples=100, centers=3, random_state=42)
df = pd.DataFrame(data, columns=['X', 'Y'])

print(df)
```

```python
plt.scatter(df['X'], df['Y'])
plt.title('Data Awal')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

linked_matrix = linkage(df, 'ward')

dendrogram(linked_matrix)
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

agglo = AgglomerativeClustering(n_clusters=3)
df['Cluster'] = agglo.fit_predict(df)

print(df)
```

```python
plt.scatter(df['X'], df['Y'], c=df['Cluster'], cmap='rainbow')
plt.title('Hasil Agglomerative Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
