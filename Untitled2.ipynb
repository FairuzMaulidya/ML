```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
from statsmodels.tsa.stattools import adfuller
```

```python
# Misalkan kita punya file CSV yang berisi data penjualan supermarket dengan kolom 'Date' dan 'Total'
data = pd.read_csv('/supermarket_sales.csv', parse_dates=['Date'], dayfirst=True)
# Menampilkan data awal untuk pengecekan
print(data.head())
```

```python
#Menyiapkan indeks sebagai tanggal
data.set_index('Date', inplace=True)
```

```python
print(data['Total'].dtype)
```

```python
# Menghapus karakter selain angka dan titik desimal
data['Total'] = data['Total'].replace(r'[^\d.]', '', regex=True)

# Mengonversi kolom 'Total' menjadi tipe numerik
data['Total'] = pd.to_numeric(data['Total'], errors='coerce')

# Cek apakah ada nilai NaN setelah konversi
print(data['Total'].isnull().sum())  # Menampilkan jumlah nilai NaN
```

```python
# Mengganti nilai NaN dengan rata-rata kolom 'Total'
data['Total'].fillna(data['Total'].mean(), inplace=True)
```

```python
print(data['Total'].dtype)  # Verifikasi tipe data
print(data['Total'].head())  # Lihat beberapa nilai pertama
```

```python
# Mengelompokkan data berdasarkan tanggal dan menjumlahkan total penjualan
# Pastikan nama kolom yang benar digunakan (misalnya 'Total' atau lainnya)
data_daily = data.groupby(data.index)['Total'].sum()

# Menampilkan data yang telah dikelompokkan
print(data_daily.head())
```

```python
print(data.describe())
```

```python
# Plot Deret Waktu
plt.figure(figsize=(10, 6))
plt.plot(data_daily.index, data_daily, marker='o', linestyle='-', color='b')
plt.title('Analisis Deret Waktu Total Penjualan Supermarket')
plt.xlabel('Tanggal')
plt.ylabel('Total Penjualan')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Melakukan Decompose deret waktu untuk melihat komponen tren, musiman, dan residual
# Menentukan periode musiman, misalnya 7 hari (untuk data harian)
result = seasonal_decompose(data_daily, model='additive', period=7)
result.plot()
plt.show()
```

```python
#Menghitung Autokorelasi (ACF) dan Partial Autokorelasi (PACF)
plt.figure(figsize=(12, 6))
```

```python
# Autokorelasi
plt.subplot(121)
plot_acf(data_daily, lags=40, ax=plt.gca())
```

```python
# Partial Autokorelasi
plt.subplot(122)
plot_pacf(data_daily, lags=40, ax=plt.gca())
plt.tight_layout()
plt.show()
```

```python
# Hitung rata-rata bergerak (moving average) dengan window 7 hari
data['Rata_Rata_Bergerak'] = data['Total'].rolling(window=7).mean()

# Plot Total dan Rata-Rata Bergerak
data[['Total', 'Rata_Rata_Bergerak']].plot(figsize=(10, 6))
plt.title('Deret Waktu Total Penjualan dengan Rata-Rata Bergerak')
plt.show()
```

```python
#Uji Stasioneritas (misalnya menggunakan Augmented Dickey-Fuller test)
result_adf = adfuller(data_daily)
print(f"ADF Statistic: {result_adf[0]}")
print(f"p-value: {result_adf[1]}")
if result_adf[1] < 0.05:
    print("Deret waktu ini stasioner.")
else:
    print("Deret waktu ini tidak stasioner.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from datetime import timedelta
```

```python
data = pd.read_csv('/supermarket_sales.csv')
print(data.head())
```

```python
# Pemrosesan dan Pembersihan Data
# Mengonversi kolom 'Date' menjadi tipe datetime
data['Date'] = pd.to_datetime(data['Date'], format='%m/%d/%Y')

# Menghapus data yang tidak relevan atau baris yang mengandung nilai kosong
data.dropna(subset=['Total'], inplace=True)
```

```python
#Mengatur Indeks Tanggal untuk Deret Waktu
# Menggunakan kolom 'Date' sebagai indeks
data.set_index('Date', inplace=True)
```

```python
# Mengelompokkan data berdasarkan hari dan menghitung total penjualan per hari
data_daily = data.resample('D')['Total'].sum()
```

```python
#Visualisasi Deret Waktu
plt.figure(figsize=(10, 6))
plt.plot(data_daily, marker='o', linestyle='-', color='b')
plt.title('Total Penjualan per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Total Penjualan')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Menggunakan dekomposisi deret waktu dengan model 'additive' dan periode musiman 30 hari
result = seasonal_decompose(data_daily, model='additive', period=30)
result.plot()
plt.tight_layout()
plt.show()
```

```python
# Plotting Autokorelasi dan Autokorelasi Parsial
plt.figure(figsize=(10, 6))
plot_acf(data_daily, lags=50)
plt.title('Autokorelasi')
plt.tight_layout()
plt.show()
```

```python
# Menghitung Autokorelasi Parsial dengan lags yang lebih kecil
plt.figure(figsize=(10, 6))
plot_pacf(data_daily, lags=30)  # Menggunakan 30 lag
plt.title('Autokorelasi Parsial')
plt.tight_layout()
plt.show()
```

```python
# Menentukan parameter ARIMA (p, d, q) berdasarkan analisis ACF dan PACF
# Misalnya kita pilih (1, 1, 1) untuk demonstrasi
model = ARIMA(data_daily, order=(1, 1, 1))

# Melatih model
model_fit = model.fit()
```

```python
# Melakukan prediksi untuk periode tertentu (misalnya 30 hari ke depan)
forecast = model_fit.forecast(steps=30)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data_daily, label='Data Asli')
plt.plot(pd.date_range(data_daily.index[-1], periods=30, freq='D'), forecast, label='Prediksi', color='red')
plt.title('Prediksi Penjualan 30 Hari ke Depan')
plt.xlabel('Tanggal')
plt.ylabel('Total Penjualan')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Menghitung error (MSE) pada model
actual = data_daily[-30:]  # Data asli selama 30 hari terakhir
mse = mean_squared_error(actual, forecast)
print(f'Mean Squared Error: {mse}')
```

```python
#Menyimpan Model ARIMA untuk Penggunaan Mendatang
model_fit.save('arima_model.pkl')
```
